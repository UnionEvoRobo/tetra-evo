# Plot Fitness Progression for One or More Runs

`plot.plot_runs([csv_paths] : list, plot_evals : bool)` takes a list of one or more string-like paths to different run.csv files and plots the fitness progression of each run against oneanother.

The last run you preformed will be sym-linked as "latest_run" so you may pass in "latest_run" instead of a path string to plot it.

The plot_evals flag controls whether to compute and plot evaluations on the x-axis instead of generations. This is done by finding the `info.json` file, which should be in the same folder as the `run.csv` file, and multiplying the generation number by the population size for each data point. This is useful to compare runs with different population sizes.

```python
import plot_utils as plot

csv_paths: list = ["latest_run", "/example/user/tetra-evo/runs/timestamp/run.csv"] # Use "latest_run" or path-like strings to run.csv files.
plot_evals = True # Whether to plot evaluations instead of generations as the x-axis.

plot.plot_runs(csv_paths, plot_evals)
```

# Plot Fitness Progression For Batch

`plot.plot_batch(batch_path : str, plot_evals : bool)` takes a path-like string to a batch folder and plots fitness over evaluations or generations for the batch with mean, min or max, and 25-75 percentile shading.

`batch_path` should be a path-like string to a batch folder. `run0`, `run1`, `run2`, ... directories should exist as subfolders with an `info.json` and `run.csv` file inside each one.

If an `info.json` file is in the same directory as the `run.csv` and the `compute_evals` flag is true, number of evaluations is computed and plotted. Otherwise, the x-axis will be generations. The `info.json` file is also relied on to correctly plot either the min or max fitness based on the `sort_reverse` flag for the run. If `sort_reverse` is True, the max will be plotted, and otherwise the min is plotted.  

```python
import plot_utils as plot

batch_path = "/example/user/tetra-evo/batches/timestamp"
plot_evals = True

plot.plot_batch(batch_path, plot_evals)
```
